```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
```

```python

# Load the iris dataset
data = load_iris()
X = data.data

```

```python

# Apply hierarchical clustering algorithm
model = AgglomerativeClustering(n_clusters=3)
model.fit(X)
labels = model.labels_

# Plot the scatter graph to visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()

```

```python
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Load iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data

# Define hierarchical clustering function
class HierarchicalClustering:
    def __init__(self, n_clusters=2, method='single'):
        self.n_clusters = n_clusters
        self.method = method
    
    def fit_predict(self, X):
        # Compute distance matrix
        dist_matrix = np.zeros((X.shape[0], X.shape[0]))
        for i in range(X.shape[0]):
            for j in range(X.shape[0]):
                dist_matrix[i,j] = np.sqrt(np.sum((X[i,:] - X[j,:])**2))
        
        # Compute linkage matrix
        Z = linkage(dist_matrix, method=self.method)
        
        # Generate dendrogram
        plt.figure(figsize=(10, 5))
        plt.title('Hierarchical Clustering Dendrogram')
        plt.xlabel('Index')
        plt.ylabel('Distance')
        dendrogram(Z, truncate_mode='lastp', p=self.n_clusters)
        plt.show()
        
        # Extract cluster labels
        clusters = np.zeros(X.shape[0], dtype=int)
        for i in range(X.shape[0]):
            for j in range(self.n_clusters):
                if i in Z[-(j+1), :2]:
                    clusters[i] = j
                    break
        
        return clusters

# Run hierarchical clustering
hc = HierarchicalClustering(n_clusters=3)
clusters = hc.fit_predict(X)
```

```python
'''This code defines a class HierarchicalClustering which takes in the number of clusters and method as input parameters. The fit_predict method takes in the dataset and returns the cluster labels. It computes the distance matrix using Euclidean distance and then computes the linkage matrix using the specified method. It then generates a dendrogram and extracts the cluster labels. Finally, it returns the cluster labels.

We can use this code to cluster the iris dataset into 3 clusters. The clusters variable will contain the cluster labels for each data point in the dataset.'''
```
