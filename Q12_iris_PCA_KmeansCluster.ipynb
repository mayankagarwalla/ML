```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python

# Load the iris dataset
data = load_iris()
X = data.data
y = data.target

```

```python
# Apply PCA to select the best 2 features
pca = PCA(n_components=2)
X_transformed = pca.fit_transform(X)

```

```python
# Plot the scatter graph using the transformed dataset
plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=y)
plt.xlabel("PCA Feature 1")
plt.ylabel("PCA Feature 2")
plt.title('Iris dataset transformed with PCA')
plt.show()
```

```python

# Apply k-means clustering algorithm to cluster the transformed dataset into 3 clusters
kmeans = KMeans(n_clusters=3)
kmeans.fit(X_transformed)
y_pred = kmeans.predict(X_transformed)

# Plot the scatter graph with the clusters
plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=y_pred)
plt.xlabel("PCA Feature 1")
plt.ylabel("PCA Feature 2")
plt.title('K-Means Clustering on Iris Dataset Transformed with PCA')
plt.show()

```

### other way

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the iris dataset
iris = load_iris()

# Perform PCA to reduce the dimensionality to 2
pca = PCA(n_components=2)
X_transformed = pca.fit_transform(iris.data)

# Plot the transformed dataset
plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=iris.target)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Iris dataset transformed with PCA')
plt.show()

# Perform K-Means clustering on the transformed dataset
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X_transformed)

# Plot the clustering results
plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200, linewidths=3, color='r')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('K-Means Clustering on Iris Dataset Transformed with PCA')
plt.show()
```
