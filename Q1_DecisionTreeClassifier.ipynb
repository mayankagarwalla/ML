```python
#Importing Useful Libraries

from sklearn import datasets,tree
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# Preparing the data-set

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
#Splitting into training and test sets

X_train,X_test,y_train,y_test = train_test_split(iris.data,iris.target,train_size = 0.80 ,test_size = 0.20,random_state = 0)
```

```python
# Fit the model

clf = DecisionTreeClassifier(criterion = 'entropy',max_depth = None,random_state = 0)
model = clf.fit(X_train,y_train)
```

```python
#Visualizing the model
text_representation = tree.export_text(clf)
print(text_representation)
```

```python
with open("decision_tree.log","w") as fout:
    fout.write(text_representation)
```

```python
fig = plt.figure(figsize = (25,15))
tree.plot_tree(clf,feature_names = iris.feature_names,class_names = iris.target_names,filled = True )
```

```python
#saving the figure
fig.savefig("decision_tree.png")
```

```python
# Test the trained model using the testing data
y_pred = clf.predict(X_test)
```

```python
# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
